```python
import numpy as np
import pandas as pd
from joblib import load
```

```python
model = load("Mining_XGBoost_cv_earlystop.joblib")
```

```python
columns = ["CMRR", "PRSUP", "depth_of_ cover", "intersection_diagonal", "mining_hight"]
random_data = pd.DataFrame({
    "CMRR": [45, 60, 75, 55, 70],
    "PRSUP": [120, 150, 110, 130, 160],
    "depth_of_ cover": [400, 500, 450, 470, 520],
    "intersection_diagonal": [8.5, 7.2, 9.0, 8.0, 10.1],
    "mining_hight": [3.0, 2.8, 3.5, 3.2, 3.1]
})
```

```python
predictions = model.predict(random_data)
print("Predicted Roof Fall Rate (for random samples):")
print(predictions)
```
